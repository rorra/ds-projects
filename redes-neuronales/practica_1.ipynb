Declarar los siguientes vectores y matrices:

$M=\begin{bmatrix}3 & -9 & 0 & 5\\
2 & -5 & -3 & 1\\
-1 & 5 & 8 & 4
\end{bmatrix}$ 

$A=\begin{bmatrix}-1 & 1 & 2\end{bmatrix}$ 
 
$B=\begin{bmatrix}\begin{array}{rrrr}
-4 & 2 & 1 & -1\end{array}\end{bmatrix}$


```python
M = np.array([[3, -9, 0, 5], [2, -5, -3, 1], [-1, 5, 8, 4]])
A = np.array([-1, 1, 2])
B = np.array([-4, 2, 1, -1])
```

Calcular los siguientes productos:

  $A\bullet M$

```python
A.dot(M)
```

$A^{\top}\bullet B$

```python
A.T.dot(B)
```

$M\bullet B^{\top}$

```python
M.dot(B.T)
```

Dadas las siguientes funciones, graficarlas en el intervalo $x\in[-2,2]$.
- $f_{1}(x)=\frac{1}{1+\exp(-x)}$
- $f_{2}(x)=\tanh(x)$
- $f_{3}(x)=\textrm{sign}(x)$
- $f_{4}(x)=f_{1}(x)*(1-f_{1}(x))$
- $f_{5}(x)=1-f_{2}(x)^{2}$

```python
x = np.linspace(-2, 2, 100)
f1 = 1/(1+np.exp(-x))
f2 = np.tanh(x)
f3 = np.sign(x)
f4 = f1*(1-f1)
f5 = 1-f2**2
```

```python
plt.plot(x, f1)
```

```python
plt.plot(x, f2)
```

```python
plt.plot(x, f3)
```

```python
plt.plot(x, f4)
```

```python
plt.plot(x, f5)
```

Graficar en una misma figura a $f_1$ con $f_4$. Hacer lo mismo con $f_2$ y $f_5$.

```python
plt.plot(x, f1, label='f1')
plt.plot(x, f4, label='f4')
plt.legend()
plt.show()
```

```python
plt.plot(x, f2, label='f2')
plt.plot(x, f5, label='f5')
plt.legend()
plt.show()
```
